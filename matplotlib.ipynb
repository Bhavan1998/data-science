```python
company=['GOOGLE', 'AMAZON', 'MSFT', 'FB']
revenue=[90,136,89,27]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.bar(company, revenue)
plt.grid()
```

```python
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company)
plt.xlabel("company")
plt.ylabel("revenue(bin)")
plt.title('US Tech Stocks')
plt.bar(ypos, revenue, label='Revenue')
plt.legend()
```

```python
company=['GOOGLE', 'AMAZON', 'MSFT', 'FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.xticks(ypos, company)
plt.xlabel("company")
plt.ylabel("revenue(bin)")
plt.title('US Tech Stocks')
plt.bar(ypos, revenue, label='Revenue')
plt.bar(ypos, profit, label='profit')
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.xlabel("company")
plt.ylabel("revenue(bin)")
plt.title('US Tech Stocks')
plt.bar(ypos-0.2, revenue,width=0.4, label='Revenue')
plt.bar(ypos+0.2, profit,width=0.4, label='profit')
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.xlabel("revenue(bin)")
plt.ylabel("company")
plt.title('US Tech Stocks')
plt.barh(ypos-0.2, revenue, label='Revenue')
plt.barh(ypos+0.2, profit, label='profit')
plt.legend()
plt.grid()
```

```python
# 1. How many patients are normal ?          80-100---Normal
# 2.How many of them are pre-diabatic ?      100-125---Pre-diabetic
# 3.How many many are diabatic ?             125 Above---Diabatic
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=3, rwidth=0.95, histtype='step')
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150], rwidth=0.95, color='black' )
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,48,69,89,79,120,112,100]
plt.xlabel('Sugar Range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['black', 'green'], label=['men', 'women'] )
plt.legend()
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,48,69,89,79,120,112,100]
plt.xlabel('Sugar Range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['black', 'green'], label=['men', 'women'], orientation='horizontal')
plt.legend()
```

```python
exp_values=[1400,600,300,410,250]
exp_labels=['Home Rent','Food','Phone/Internet Bill', 'car', 'other utilities']
```

```python
plt.pie(exp_values, labels=exp_labels)
plt.axis('equal')
```

```python
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0,0.2,0.2,0.4,0.7], startangle=180)
plt.axis('equal')
plt.show()
```
