```python
def people_cell(cell):
    if cell=='n.a.':
        return 'samul'
    return cell
def cell(cell):
    if cell=="not available":
         return None
    return cell
def rev_cell(cell):
    if cell=='-1':
        return 66
    return cell

def tic_cell(cell):
    if cell=='GOOGLE':
        return 'apple'
    return cell

df=pd.read_excel('stock market.xlsx', converters={
    'people': people_cell,
    'eps': cell, 
    'revenue': rev_cell,
    'tickers': tic_cell
})
```

```python
df
```
